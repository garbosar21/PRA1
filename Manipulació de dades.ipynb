```python
import pandas as pd
```

# 1. Tratamiento Embassaments

```python
df = pd.read_csv('Quantitat_d_aigua_als_embassaments_de_les_Conques_Internes_de_Catalunya.csv')
```

```python
df['fecha'] = pd.to_datetime(df['Dia'], format='%d/%m/%Y')

# Crear nuevas columnas 'mes' y 'año'
df['mes'] = df['fecha'].dt.month
df['año'] = df['fecha'].dt.year
df['dia'] = df['fecha'].dt.day
```

```python
dicc_estacio = {'Embassament de Siurana (Cornudella de Montsant)': 'Siruana','Embassament de la Baells (Cercs)': 'Baells',
               'Embassament de Sant Ponç (Clariana de Cardener)': 'Sant Ponç', 'Embassament de Foix (Castellet i la Gornal)': 'Foix',
               'Embassament de Susqueda (Osor)': 'Susqueda', 'Embassament de Riudecanyes': 'Riudecanyes',
               'Embassament de Darnius Boadella (Darnius)': 'Darnius Boadella', 'Embassament de Sau (Vilanova de Sau)': 'Sau', 
               'Embassament de la Llosa del Cavall (Navès)': 'Llosa del Cavall'}
df['Estació'] = df['Estació'].replace(dicc_estacio)
```

```python
df_merged = df[['dia', 'mes', 'año']]
llista_est = df['Estació'].unique()
for estacio in df['Estació'].unique():
    df_merged = pd.merge(df_merged, df[df['Estació']==estacio], on=['dia', 'mes', 'año'], suffixes=['', f'_{estacio}'])
```

```python
df_merged.columns
```

```python
df_grouped = df.groupby(['dia',  'mes', 'año'], as_index=False).agg(Percentatge_volum_embassat_cat=('Percentatge volum embassat (%)', 'mean'))
df_grouped
```

```python
df_final = pd.merge(df_merged, df_grouped, on=['dia',  'mes', 'año'])
df_final
```

# 2. Tratamiento Meteorológiques

```python
df1_sum = pd.read_csv('Dades_meteorol_giques_de_la_XEMA.csv')
df1_sum.rename(columns={'VALOR_LECTURA': 'Total'}, inplace=True)
df1_sum
```

```python
df1_avg = pd.read_csv('Dades_meteorol_giques_de_la_XEMA_avg.csv')
df1_avg.rename(columns={'VALOR_LECTURA': 'Promig'}, inplace=True)
df1_avg
```

```python
df1 = pd.merge(df1_sum, df1_avg, on=['DATA_LECTURA', 'CODI_VARIABLE'], how='inner')
df1
```

```python
df1['fecha'] = pd.to_datetime(df1['DATA_LECTURA'], format='%d/%m/%Y %I:%M:%S %p')

# Crear nuevas columnas 'mes' y 'año'
df1['mes'] = df1['fecha'].dt.month
df1['año'] = df1['fecha'].dt.year
df1['dia'] = df1['fecha'].dt.day
```

```python
dicc_var= {35: 'Precipitació', 40: 'Temp. Màxima', 42: 'Temp. Mínima'}
```

```python
df1['CODI_VARIABLE'] = df1['CODI_VARIABLE'].replace(dicc_var)
df1
```

```python
df1_grouped = df1.groupby(['CODI_VARIABLE', 'mes', 'año', 'dia'], as_index=False).agg(Valor_Total=('Total', 'sum'), Valor_Promig=('Promig', 'mean'))
df1_preci = pd.DataFrame()
df1_max = pd.DataFrame()
df1_min = pd.DataFrame()

df1_preci[['dia', 'mes', 'año', 'Valor_precipitació']] = df1_grouped[df1_grouped['CODI_VARIABLE']=='Precipitació'][['dia', 'mes', 'año', 'Valor_Total']]
df1_max[['dia', 'mes', 'año', 'Valor_temp_max']] = df1_grouped[df1_grouped['CODI_VARIABLE']=='Temp. Màxima'][['dia', 'mes', 'año', 'Valor_Promig']]
df1_min[['dia', 'mes', 'año', 'Valor_temp_min']] = df1_grouped[df1_grouped['CODI_VARIABLE']=='Temp. Mínima'][['dia', 'mes', 'año', 'Valor_Promig']]
df1_final = pd.merge(df1_preci, df1_max, on=['dia', 'mes', 'año'])
df1_final = pd.merge(df1_final, df1_min, on=['dia', 'mes', 'año'])
```

```python
df1_final
```

```python
df_definitivo=pd.merge(df_final, df1_final, on=['dia', 'mes', 'año'])
df_definitivo.drop_duplicates(inplace=True)
```

```python
df_definitivo.rename(columns={'Dia': 'Dia_Siruana', 'Nivell absolut (msnm)': 'Nivell absolut (msnm)_Siruana', 
                              'Percentatge volum embassat (%)': 'Percentatge volum embassat (%)_Siruana',
                             'Volum embassat (hm3)': 'Volum embassat (hm3)_Siruana', 'fecha': 'fecha_Siruana', 'Estació': 'Estació_Siruana'}, inplace=True)
for est in llista_est:
    df_definitivo.drop([f'Estació_{est}', f'Dia_{est}', f'fecha_{est}'], axis=1, inplace=True)
```
